```
from matplotlib import image
from os import listdir

#Load all images in DATASET/TRAIN/R  (images Recyclable items)
tRecycled = list()
for filename in listdir('R'):
	#Loading image by image in folder R
	img_data = image.imread('R/' + filename)
	#Storing
	tRecycled.append(img_data)
	print('> loaded %s %s' % (filename, img_data.shape))
  
```

```
# load all images into a list, from DATASET/TRAIN/O (organic images)
tOrganic = list()
for filename in listdir('O'):
	#Loading image by image in folder O
	img_data = image.imread('O/' + filename)
	#Storing
	tOrganic.append(img_data)
	print('> loaded %s %s' % (filename, img_data.shape))
```

```
#Saving all training records into npy files for later manipulation (SEE CONVNET3 Colab file)
import numpy as np

#Saving records from Recycled Training folder
np.save('FtRecycled', np.asarray(tRecycled), allow_pickle=True)

#Saving records from Organic Training folder
np.save('FtOrganic', np.asarray(tOrganic), allow_pickle=True)
```
